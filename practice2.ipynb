```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
df.head()
```

```python
#Task 2(Exploration of dataset)

print(f"number of rows:{df.shape[0]}")
print(f"number of columns:{df.shape[1]}")

print(f"\nData types of each column:\n{df.dtypes}")
print(f"\nMissing values:\n{df.isnull().sum()}")

print("\nNumerical features:")
numerical=df.select_dtypes(include=['int64','float64']).columns.tolist()
print(numerical)

print("\nCategorical features:")
categorical=df.select_dtypes(include=['object', 'str']).columns.tolist()
print(categorical)
```

```python
#Task 3 (Data Cleaning)
#task D - remove rows with missing values

print("Missing values before cleaning:")
print(df.isnull().sum())
print(f"Total rows before cleaning: {df.shape[0]}")
rows_before=df.shape[0]

#removing irrelevant columns
df=df.drop(columns=['Name', 'Ticket', 'Cabin'])
print("\nIrrelevant columns removed: Name, Ticket, Cabin")

#Removing rows with missing values
df=df.dropna()

print("\nMissing values after cleaning:")
print(df.isnull().sum())
rows_after=df.shape[0]

print(f"\nTotal rows after cleaning:{rows_after}")
print(f"\nRows removed:{rows_before - rows_after}")
```

```python
#Task 4 (Data transformation)

df['Sex']=df['Sex'].map({'male':0, 'female':1})
df['Embarked']=df['Embarked'].map({'S': 0, 'C': 1, 'Q': 2})
print(df.dtypes)
df.head(10)
```

```python
#Task 5 (Feature Selection)
 
y=df['Survived']
print("Target variable: Survived")
print(y.value_counts())
print("\n0 is did not survived\n1 is survived")

#select at least 4 features
X=df[['Pclass', 'Sex', 'Age', 'Fare']]
print(f"\nSelected Features: \n{X.head()}")
```

Selected features:

Pclass - passenger class affects survival

Sex - strong correlation with survival

Age - age impacts survival probability

Fare - reflects socio-economic status

```python
#Task 6(Final Dataset)

print("Missing values in final dataset:")
print(df.isnull().sum())
print(f"Total missing values: {df.isnull().sum().sum()}")

print(f"\nData types of final dataset: \n{df.dtypes}")

print(f"\nFinal dataset size: {df.shape[0]} rows, {df.shape[1]} features")
print(f"Final dataset: \n{df.head()}")
```
