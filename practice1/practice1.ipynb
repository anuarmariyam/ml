```python
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
#Task 1: dataset loading
iris = load_iris() 

df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["species"]=iris.target
df["species_name"]=df["species"].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

print(df.head())
# print(iris.target_names)
```

```python
#Task 2: dataset exploration
#number of rows, columns, features
print("\n Dataset size")
print(f"number of rows: {df.shape[0]}")
print(f"number of columns: {df.shape[1]}")

num_features=len(iris.feature_names)
print(f"number of features: {num_features}")
print(f"feature names: {iris.feature_names}")

#types of data are contained in the table
print("\n Data types")
print(df.dtypes)

#missing values
print("\n Missing values")
print(df.isnull().sum())

#basic statistics
print("\n Basic statistics")
print(df.describe().round(2))
```

```python
#Task 3: vizualization
#task D - pairplot

sns.pairplot(df, hue="species_name", palette="Set1")

plt.suptitle("Task D - Pairplot: Relationships between all features", y=1.02)
plt.show()
```

### Task 3: Analysis
Looking at the pairplot, Iris Setosa is clearly separated from the other two species in all graphs, especially by petal length and petal width. Iris Versicolor and Iris Virginica are close to each other but still mostly separable. The diagonal histograms confirm that petal length and petal width have the most distinct distributions across species.

### Task 4: Conclusion

The most informative features in the Iris dataset are petal length and petal width, as they show the clearest difference between the three species. 
It is possible to visually separate the species — Iris Setosa is completely separated from Versicolor and Virginica, while the other two species have only a small overlapping zone.
Petal length and petal width are strongly correlated with each other, meaning that flowers with longer petals always tend to have wider petals as well.
Overall, even a simple visual analysis is enough to distinguish between the iris species with high confidence.
